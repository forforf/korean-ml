## Train Model and Save: SVC (Support Vector Classifier)

We wrap the model in a pipeline as we need to do some transformations (creating windowed features for X).

```python



sliding_window = SlidingWindow()
disp.obj(sliding_window, label='SlidingWindow')


# Models and processing
model_id = 'logr'
logr = LogisticRegression()
# Using sklearn pipeline creates a problem with comparing saved model vs current model since the model instance is
# returned with `pipe.get_params`, rather than the model parameters. I need to figure out a way to get around that
# in order to prevent unnecessary model updates.
# from transformers import SlidingWindow
steps = [('sw', sliding_window), (model_id, logr)]
pw = PipeWrap(steps)
```

```python
def get_boolean_delta(a1):
    # TODO: Vectorize
    d = np.zeros(len(a1))
    for i, el in enumerate(a1):
        if i == 0:
            continue
        else:
            if a1[i] > a1[i-1]:
                d[i] = 1
            elif a1[i] < a1[i-1]:
                d[i] = -1
            else:
                # el is already 0
                pass
    return d

def get_transitions(ba):
    return np.where(ba == 1)[0], np.where(ba == -1)[0]

def get_transition_dists(ta1, ta2):
    """
    TODO: There are a lot of ways to optimize this. For example:
    1) We don't need the full matrix, just the minimums, so there might be a way to skip creating the
       full distance matrix
    2) When expanding a vector into a matrix, it'd probably be better to choose the smaller vector.
    Get the distances between transitions
    :param ta1: First vector (M,) with transition indices
    :param ta2: Second vector (N,) with transition indices
    :return: Distance Matrix (M,N) for all transitions
    """
    ta1_m = np.repeat(ta1, ta2.shape[0]).reshape(ta1.shape[0], ta2.shape[0])
    return abs(ta1_m - ta2)

def dist_penalty(dm):
    shortest_dists = np.min(dm, axis=0)
    return np.sum(shortest_dists)

def event_ticker(a):
    direction = -1
    counter = 0
    ctr = np.zeros(len(a))
    for i, el in enumerate(a):
        # reset counter if event occurred and set direction
        if el != 0:
            direction = el
        else:
            counter += direction

        ctr[i] = counter
    return ctr

def get_penalty(a1, a2):
    bd = get_boolean_delta
    t = get_transitions
    td = get_transition_dists
    d = dist_penalty
    a1ft, a1tf = t(bd(a1))
    a2ft, a2tf = t(bd(a2))
    ft_penalty = d(td(a1ft, a2ft))
    tf_penalty = d(td(a1tf, a2tf))
    return ft_penalty + tf_penalty

from sklearn.metrics import r2_score
def ticker_error(a1, a2):
    b1 = get_boolean_delta(a1)
    b2 = get_boolean_delta(a2)
    ev1 = event_ticker(b1)
    ev2 = event_ticker(b2)
    return r2_score(ev1, ev2)

```

```python
from scorers import ConvVar

dist_pen = make_scorer(get_penalty, greater_is_better=False)
convvar_score = make_scorer(ConvVar.score)

# Good SW Parameters that keep appearing
# 148:0.75
sw = [128] #np.logspace(6.0, 11.0, num=8, base=2.0).astype(int) # np.linspace(958-300, 958+300, num=4).astype(int) #

off_perc = [0.65] #np.linspace(0.5, 0.95, num=8)
print(f'{sw}:{off_perc}')
p = {
    'sw__window': sw, #[512, 1024, 2056, 4094],
    'sw__offset_percent': off_perc,
    'logr__C': [2.0],
    'logr__tol': [0.01],
    'logr__max_iter': [1000]
}

tscv = TimeSeriesSplit(n_splits=5)
search = GridSearchCV(pw.pipe, p, n_jobs=-1, cv=tscv, scoring=convvar_score, verbose=4)
search.fit(X_train, y_train)
print(f'Best parameter (CV score={search.best_score_:.3f}):')
print(search.best_params_)

pw.pipe = search.best_estimator_
```

## Evaluate Training

Evaluate the training data by seeing if our trained model will predict close to the data it was trained on.

Validation will come later

This is just to make sure our Model trained correctly. It might have overfit, but we don't care yet.

```python
# TODO: Move these into a class
def pred_threshold(pred_vals, thresh=0.5):
    return np.where(pred_vals > thresh, True, False)

y_train_pred = saved_wrap.predict(X_train)
```

```python
shared_params = joblib.load(fv_shared_params.get_latest_path())
# remove kss_ids from params just to keep things a bit cleaner
kss_ids = shared_params['kss']['trn_ids']
print(kss_ids)
audio_params = shared_params['audio']
# iterate over each kss_id, find the audio file, load it and return it's squeezed rms
x_rms = np.concatenate([Audio(KssDf(id).audio_path, **audio_params).rms.squeeze() for id in kss_ids])

sw_mean_offset = sliding_window.window-sliding_window.offset
sw_offset = 0

def align(y, offset):
    return np.pad(y, (offset, 0), 'minimum')

def delta(y1, y2, y1_offset, y2_offset):
    min_len = min(len(y1), len(y2))
    return np.logical_xor(align(y1, y1_offset)[0:min_len], align(y2, y2_offset)[0:min_len])

def speech_base_delta(y, offset):
    return delta(y, y_train, offset, 0)
```
