```python
from nbtplt.core import NbTpltChain

# Usage
# nbchain = NbtpltChain(generated_doc_path_prefix, template_nb_path, source_nb_paths)
# nbchain.generate() # to create templated notebook
#
# Or you can do it without interim variables
# NbtpltChain(generated_doc_path_prefix, template_nb_path, source_nb_paths).generate()

SRC = './template-src'
GEN = './generated-nb'
SK = 'sklearn'
SK_PRE = f'{SK}-'
KERAS = 'keras'
KERAS_PRE = f'{KERAS}-'
MODEL = 'model'
MODEL_PRE = f'{MODEL}-'
```

```python
def get_tplt_name(basename):
    return f'{basename}.nbtplt.ipynb'

def get_src_name(basename):
    return f'{SRC}/{get_tplt_name(basename)}'

def get_gen_name(prefix, basename):
    # return f'{GEN}/{SK_PRE}{get_tplt_name(basename)}'
    return f'{prefix}{get_tplt_name(basename)}'

sklearn_names = ['logr', 'svr-rbf', 'tree']
keras_names = ['cnn']
```

```python
sklearn_template_inputs = [get_src_name(n) for n in sklearn_names]

print(sklearn_template_inputs)

sklearn_chain = NbTpltChain(f'{GEN}/{SK_PRE}',
                            f'{SRC}/{SK}.nbtplt-proxy.ipynb',
                            sklearn_template_inputs )
sklearn_chain.generate()
```

```python
keras_template_inputs = [get_src_name(n) for n in keras_names]
keras_chain = NbTpltChain(f'{GEN}/{KERAS_PRE}',
                            f'{SRC}/{KERAS}.nbtplt-proxy.ipynb',
                            keras_template_inputs)

print(keras_template_inputs)
keras_chain.generate()
```

```python
# Uses files generated by the above chains
sk_gen_pre = f'{GEN}/{SK_PRE}'
sklean_model_inputs = [get_gen_name(sk_gen_pre, n) for n in sklearn_names]

keras_gen_pre = f'{GEN}/{KERAS_PRE}'
keras_model_inputs = [get_gen_name(keras_gen_pre, n) for n in keras_names]

model_template_inputs = sklean_model_inputs + keras_model_inputs
print(model_template_inputs)
```

```python
# Generate model notebooks

model_chain = NbTpltChain(f'{GEN}/{MODEL_PRE}',
                          f'{SRC}/{MODEL}.nbtplt.ipynb',
                          model_template_inputs)

model_chain.generate()

```
